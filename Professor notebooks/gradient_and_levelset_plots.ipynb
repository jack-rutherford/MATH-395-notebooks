```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
# Generating data
x = np.linspace(-3, 3, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y) 
Z = 2*X**2 + Y**4 - 4*X*Y # function formula

# Defining contour levels
levels = np.linspace(-3, 3, 101)

# Creating contour plot
contour = plt.contour(X, Y, Z, cmap=cm.coolwarm,levels = levels)

# Adding colorbar
plt.colorbar(contour, label='f-values')

# Adding labels and title
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('$f(x_1,x_2)=2x_1^2 + x_2^4 - 4 x_1 x_2$')

# Displaying the plot
plt.show()
```

```python
# Directional vectors 

# around (-1,-1)
x = np.linspace(-1.5, -0.5, 31)
y = np.linspace(-1.5, -0.5, 21)

# around (0,0)
#x = np.linspace(-0.3, 0.3, 31)
#y = np.linspace(-0.2, 0.2, 21)

X, Y = np.meshgrid(x, y)

partial_x = 4*X - 4*Y
partial_y = 4*Y**3 -4*X
  
# Plotting Vector Field with QUIVER 
plt.quiver(X, Y, partial_x, partial_y, color='b') 
#plt.title('Vector Field') 

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('gradient field')

# Show plot with grid 
plt.grid() 
plt.show() 
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
x = np.linspace(-1.5, 1.5, 31)
y = np.linspace(-1, 1, 21)
X, Y = np.meshgrid(X, Y)

Z = 2*X**2 + Y**4 - 4*X*Y # function formula

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
#ax.set_zlim(0, 0.3)
#ax.zaxis.set_major_locator(LinearLocator(11))

# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

#plt.axis('square')
plt.show()
```
